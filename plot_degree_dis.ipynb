```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
```

```python
ba = nx.barabasi_albert_graph(4000, 2, seed=666)
nd = list(ba.degree())
degree = [ndd[1] for ndd in nd]
```

```python
count, bins, _ = plt.hist(degree, bins=200)
count
# bins
# plt.xlim(0, 10)
```

```python
degree_count = [0 for i in range(21)]

x = [i for i in range(0, 21)]

for d in degree:
    if d < 21:
        degree_count[d] += 1
        
plt.figure(figsize=(10, 8))
        
plt.bar(x, degree_count, width=0.8)

for xx in x:
    if xx > 1:
        plt.text(xx-0.4, degree_count[xx]+20, str(degree_count[xx]))


plt.xlim(0, 21)
plt.xticks([i for i in range(0, 21)])
plt.xlabel('degree')
plt.ylabel('number')
plt.title('Degree Distribution for degree under 21')
plt.show()
```

```python
ba = nx.barabasi_albert_graph(4000, 2, seed=666)
nd = list(ba.degree())
degree = [ndd[1] for ndd in nd]
```

```python
count, bins, _ = plt.hist(degree, bins=200)
count
# bins
# plt.xlim(0, 10)
```

```python
degree_count = [0 for i in range(21)]

x = [i for i in range(0, 21)]

for d in degree:
    if d < 21:
        degree_count[d] += 1
        
plt.figure(figsize=(5, 3.5))
        
plt.bar(x, degree_count, width=0.8)

for xx in x:
    if xx > 1:
        plt.text(xx-0.7, degree_count[xx]+20, str(degree_count[xx]), size=8)


plt.xlim(0, 21)
plt.xticks([i for i in range(0, 21)])
plt.xlabel('degree', size=12)
plt.ylabel('number', size=12)
plt.title('Degree Distribution for degree k < 21', size=13)
plt.savefig('degree_dis.png', bbox_inches='tight')
plt.show()
```

```python
cum_dis = np.cumsum(degree_count)/4000

plt.figure(figsize=(5, 3.5))

plt.plot(x, cum_dis)

plt.plot([2, 2], [0, cum_dis[2]], 'r', linestyle='--', label='k* = 2')
plt.plot([0, 2], [cum_dis[2], cum_dis[2]], 'r', linestyle='--')

plt.plot([3, 3], [0, cum_dis[3]], 'purple', linestyle='--', label='k* = 3')
plt.plot([0, 3], [cum_dis[3], cum_dis[3]], 'purple', linestyle='--')

plt.text(-2.5, cum_dis[2], str(cum_dis[2] * 100)[:4]+'%', color='r', size=15)
plt.text(-2.5, cum_dis[3], str(cum_dis[3] * 100)[:4]+'%', color='purple', size=15)


plt.xlim(0, 21)
plt.ylim(0, 1)
plt.xticks([i for i in range(0, 21)])
plt.yticks(np.linspace(0, 1, 6, endpoint=True), [str(i)+'%' for i in range(0, 101, 20)])
plt.xlabel('degree',size=12)
plt.ylabel('Percentage', size=12)
plt.title('Cumulative Degree Distribution for k < 21', size=13)

plt.legend()
plt.savefig('degree_cum_dis.png', bbox_inches='tight')
plt.show()
```
