```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('part1 data.csv')
```

```python
cols = data.columns
x = [1+i for i in range(data.shape[0])]
plt.plot(x, data[cols[0]], label=cols[0])
for c in range(1, len(cols)):
    plt.plot(x, data[cols[c]].to_numpy()/4000, label=cols[c])
plt.show()
```
