```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('part1_data.csv')
```

```python
N = 4000
bs = np.arange(1, 5.6, 0.1)
plt.figure(figsize=(10, 8))
plt.plot(bs, df['<c>'], 'o-', label='<c>')
plt.plot(bs, df['PC'].to_numpy() / N, '^-', label='PC')
plt.plot(bs, df['PD'].to_numpy() / N, 's-', label='PD')
plt.plot(bs, df['F'].to_numpy() / N, '*-', label='F')
plt.legend(edgecolor='w')
plt.xlabel('b')
plt.savefig('pic_part1.png')
plt.show()
```
